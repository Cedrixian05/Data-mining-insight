```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import f_oneway
import warnings
warnings.filterwarnings('ignore')

```

```python
internet = pd.read_csv('Internet-Demo.csv')
```

<h1> Categorical Data Fields </h1>

```python
internet['Country Name'] = internet['Country Name'].astype('category')
internet['Country Code'] = internet['Country Code'].astype('category')
internet['Income Group'] = internet['Income Group'].astype('category')
```

```python
internet.dtypes
```

<h1> Question 1 : Histogram Distribution of Birth Rates </h1>

```python
plt.figure(figsize=(12, 10))
sns.histplot(internet["Birth rate"], bins=5, kde=True, color="purple")
plt.xlabel("Birth Rate", color="darkblue")
plt.ylabel("Frequency", color="darkblue")
plt.title("Distribution of Birth Rates", color="darkred")
plt.grid(True)
plt.show()
```

<h1> Question 2 : Top 5 Countries with Highest Internet Usage </h1>

```python
top_5_internet_usage = internet[["Country Name", "Internet users"]].sort_values(by="Internet users", ascending=False).head(5)
```

```python
top_5_internet_usage 
```

<h1> Question 3 :  Count of Countries by Income Group </h1>

```python
income_count = internet["Income Group"].value_counts()
```

```python
income_count
```

```python
plt.figure(figsize=(8, 6))
plt.bar(income_count.index, income_count.values, color='red')
plt.xlabel('Income Group')
plt.ylabel('Count of Countries')
plt.title('Count of Countries by Income Group')
plt.tight_layout()
plt.show()
```

<h1> Question 4 :Box Plot of Internet Usage </h1>

```python
plt.figure(figsize=(10, 8))
sns.boxplot(y=internet["Internet users"], color="teal")
plt.ylabel("Internet Users (%)", color="darkblue")
plt.title("Box Plot of Internet Usage", color="darkred")
plt.show()
```

<h1> Question 5 : Countries with No Internet Access </h1>

```python
display(internet.head())
no_internet = internet[internet['Internet users'] == 0]['Country Name']
print(no_internet)
```

<h1> Question 6 : Average Birth Rate </h1>

```python
avg_birth_rate = internet["Birth rate"].mean()
print(f"Average Birth Rate: {avg_birth_rate:.2f}")
```

<h1> Question 7 : Scatter Plot: Birth Rate vs. Internet Usage </h1>

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x=internet["Birth rate"], y=internet["Internet users"], color="RED")
plt.xlabel("Birth Rate")
plt.ylabel("Internet Users (%)")
plt.title("Scatter Plot: Birth Rate vs. Internet Usage", color="darkred")
plt.grid(True)
plt.show()
```

<h1> Question 8 : Countries with Birth Rate Above 40 </h1>

```python
high_birth_rate = internet[internet["Birth rate"] > 40]
print("Countries with Birth Rate Above 40:")
print(high_birth_rate[["Country Name", "Birth rate"]])
```

<h1> Question 9 : Pie Chart of Income Groups </h1>

```python
plt.figure(figsize=(12, 10))
internet["Income Group"].value_counts().plot.pie(autopct='%1.1f%%', colors=["red", "blue", "green"], textprops={'color': "darkblue"})
plt.title("Proportion of Countries by Income Group", color="darkred")
plt.ylabel("")
plt.show()
```

<h1> Question 10 : Correlation Between Birth Rate and Internet Usage </h1>

```python
correlation = internet["Birth rate"].corr(internet["Internet users"])
```
