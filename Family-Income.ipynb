<h1>Data Preperation</h1>

```python
import numpy as np
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import warnings 
warnings.filterwarnings('ignore')
```

<h1>Creating Data Frame</h1>

```python
df = pd.read_csv('dataset\\Family Income and Expenditure.csv')
```

```python
df
```

```python
df.columns
```

<h1>Checking for missing Values</h1>

```python
df.isnull().sum()
```

<h1>Filling up Missing Values in the Data Frame</h1>

```python
df['Household Head Occupation'] = df['Household Head Occupation'].fillna(df['Household Head Occupation'].mode()[0])
df['Household Head Class of Worker'] = df['Household Head Class of Worker'].fillna(df['Household Head Class of Worker'].mode()[0])
df['Toilet Facilities'] = df['Toilet Facilities'].fillna(df['Toilet Facilities'].mode()[0])
```

<h1>Insight 1: What is the distribution of the main sources of income among households?</h1>

```python
plt.figure()
df["Main Source of Income"].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette("pastel"))
plt.title("Distribution of Main Source of Income")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

<h1>Wage/Salaries: The largest portion of income comes from wage/salaries , accounting for 49.1% of the total. This indicates that the majority of individuals rely on employment-based income as their primary source of earnings.</h1>

<h1>Other Sources of Income: The second-largest source is 26.1% of income comes from other sources , which could include investments, pensions, government assistance, or other miscellaneous income streams.</h1>

<h1>Entrepreneurial Activities: The remaining is entrepreneurial activities , contributing 24.8% of the total income. This suggests that a significant number of people derive income from self-employment, small businesses, or other entrepreneurial ventures.</h1>

<h1>Insight 2: What is the average total household income by type of household?</h1>

```python

plt.figure(figsize=(10, 6))
sns.barplot(x='Type of Household', y='Total Household Income', data=df)
plt.title('Average Total Income by Household Type')
plt.xlabel('Type of Household')
plt.ylabel('Average Total Household Income')
plt.show()
```

<h1>Insight 2: The bar chart shows that households consisting of two or more nonrelated persons/members have the highest average total income 360,000, followed by extended families 300,000  and single-family households 220,000. This indicates that income levels are influenced by household composition</h1>

<h1> Insight 3: Average income and food expenditure per region </h1>


```python
region_avg = df.groupby('Region')[['Total Household Income', 'Total Food Expenditure']].mean()
region_avg.columns = ['Average Income', 'Average Food Spending']
region_avg.plot(kind='bar', color=['orange', 'lightblue'], figsize=(14, 8))
plt.title('Average Income vs Food Spending by Region')
plt.ylabel('Amount (PHP)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

<h1> Insight 3: Across The regions, households with higher incomes like NCR spend more money on food, yet food consistently consumes roughly one-quarter to one-third of income everywhere. so clearly the NCR earns and spends the most.</h1>

<h1>  Insight 4: Average Number of Children by Age Group per Region </h1>

```python
grouped = df.groupby('Region')[['Members with age less than 5 year old','Members with age 5 - 17 years old']].mean()
grouped.plot(kind='bar', stacked=True, figsize=(14, 8), color=['#FF9999', '#66B2FF'])
plt.title('Average Number of Children by Age Group per Region')
plt.xlabel('Region')
plt.ylabel('Average Count')
plt.xticks(rotation=90)
plt.tight_layout()
plt.legend(title='Age Group')
plt.show()
```

<h1>  Insight 4: This stacked bar chart reveals significant regional differences in the average number of children per household, with regions like ARMM showing higher child populations compared to NCR. It also highlights that younger children under 5 are more evenly distributed across regions than older children 5â€“17 years. </h1>

<h1> Insight 5: Histogram of Household Head Age </h1>


```python
plt.figure(figsize=(10, 6))
sns.histplot(df["Household Head Age"], color='green', bins=30, kde=True)
plt.title("Distribution of Household Head Age")
plt.xlabel("Age")
plt.ylabel("Number of Households")
plt.tight_layout()
plt.show()
```

<h1>Insight 8: Which food categories have the highest and lowest average household expenditures?</h1>

<h1>Insight 12: What are the most common sources of water supply among households?</h1>

```python
water_supply_distribution = df['Main Source of Water Supply'].value_counts()
water_supply_distribution.plot(kind='bar', figsize=(10, 6))
plt.title('Main Sources of Water Supply')
plt.xlabel('Source of Water Supply')
plt.ylabel('Number of Households')
plt.show()
```

<h1> Insight 13: Correlation Between Major Expenditure Categories  </h1>


```python
expenditure = [
    "Total Food Expenditure", "Meat Expenditure", "Total Fish and  marine products Expenditure",
    "Fruit Expenditure", "Vegetables Expenditure", "Transportation Expenditure",
    "Education Expenditure", "Medical Care Expenditure"
]
plt.figure(figsize=(10, 8))
sns.heatmap(df[expenditure].corr(), annot=True, cmap="magma", fmt=".2f")
plt.title("Correlation Between Major Expenditure Categories")
plt.tight_layout()
plt.show()
```
