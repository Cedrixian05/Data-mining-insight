```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import folium
from folium.plugins import HeatMap

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Global_terrorist.csv', encoding='ISO-8859-1')
```

```python
df
```

```python
filter_df = df[df['region_txt'] == 'Middle East & North Africa']

#filter the year
filter_df = filter_df[filter_df['iyear'] >= 2000]

#filter the attack type
filter_df = filter_df[filter_df['attacktype1_txt'] == 'Bombing/Explosion']

filter_df = filter_df[filter_df['nkill'] >= 10]

print(f'Filtered dataset size: {len(filter_df)} rows')
```

```python
filter_df
```

<h1> Question 1 </h1>
<h2> Which Region has the highest Bombing attack </h2>

```python
filtered_df = df[(df['attacktype1_txt'] == 'Bombing/Explosion') & (df['iyear'] >= 2000)]

aggregated_df = filtered_df.groupby(['latitude', 'longitude']).size().reset_index(name='incident_count')

aggregated_df['normalized_weight'] = (aggregated_df['incident_count'] - aggregated_df['incident_count'].min()) / \
                                     (aggregated_df['incident_count'].max() - aggregated_df['incident_count'].min())

heat_data = aggregated_df[['latitude', 'longitude', 'normalized_weight']].values.tolist()

m = folium.Map(location = [30, 40], zoom_start = 4)

HeatMap(heat_data).add_to(m)

m.save('terrorism_heatmap.html')
m
```

```python
bombing_explosion_df = df[df['attacktype1_txt'] == 'Bombing/Explosion']

region_bombing_counts = bombing_explosion_df.groupby('region_txt').size().reset_index(name='bombing_count')

highest_bombing_region = region_bombing_counts.sort_values(by='bombing_count', ascending=False).iloc[0]

print(highest_bombing_region)
```

<h1> Question 3 </h1>
<h2> Casualty Analysis </h2>
<h1> Which attack type and region have the highest casualties </h1>

```python
casualty_df = df[(df['iyear'] >= 2010) & (df['weaptype1_txt'] == 'Firearms')]

casualty_df['total_casualties'] = casualty_df['nkill'] + casualty_df['nwound']

casualty_df = casualty_df[casualty_df['total_casualties'] > 0]

aggregated_casualty = casualty_df.groupby(['latitude', 'longitude']).agg({'total_casualties': 'sum'}).reset_index()

aggregated_casualty['normalized_weight'] = (aggregated_casualty['total_casualties'] - aggregated_casualty['total_casualties'].min()) / \
                                            (aggregated_casualty['total_casualties'].max() - aggregated_casualty['total_casualties'].min())

heat_data = aggregated_casualty[['latitude', 'longitude', 'normalized_weight']].values.tolist()

casualty_insight = folium.Map(location=[30, 40], zoom_start=4)

HeatMap(heat_data).add_to(casualty_insight)

casualty_insight.save('casualty_heatmap_firearms.html')

casualty_insight
```

<h1> Insigth 3.1</h1>

<h2>The region with the highest casualties is Middle East & North Africa, with a total of 329,807 casualties.</h2>

```python
df['total_casualties'] = df['nkill'] + df['nwound']

region_casualties = df.groupby('region_txt')['total_casualties'].sum().reset_index()

region_casualties_sorted = region_casualties.sort_values(by='total_casualties', ascending=False)

highest_casualty_region = region_casualties_sorted.iloc[0]
print(f"The region with the highest casualties is {highest_casualty_region['region_txt']} with {highest_casualty_region['total_casualties']} total casualties.")
```
